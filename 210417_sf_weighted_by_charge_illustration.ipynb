```python
from logp.sf_behler.sf import get_charge
from rdkit import Chem
from rdkit.Chem.Draw import SimilarityMaps
from logp.sf_behler.g3D import XYZ_dict
from logp.sf_behler.sf import sum_G1
from logp.sf_behler.sf import sum_G2
import matplotlib.pyplot as plt
```

```python
smile = 'O'
mol = Chem.MolFromSmiles(smile)
mH = Chem.AddHs(mol)
charge = get_charge(mH)
```

```python
print(charge)
```

# Visualize the Gasteiger Partial charges 

- charge q

```python
fig1 = SimilarityMaps.GetSimilarityMapFromWeights(mH, charge, colorMap='jet',contourLines=10,scale=-1)
```

```python
XYZ, elems, elemsidx = XYZ_dict(mH); XYZ
```

## Calculate the Symmetry Function G1 with the weights set to be 1.0

- G1 scaled by 1 $eta$=2, and $Rs$=0

```python
G1_2_0 = sum_G1(XYZ,eta=2,Rs=0,weights=None);G1_2_0
```

```python
fig2 = SimilarityMaps.GetSimilarityMapFromWeights(mH, G1_2_0, colorMap='jet',contourLines=10,scale=-1)
```

- G1 scaled by 1 $eta$=5, and $Rs$=0

```python
G1_5_0 = sum_G1(XYZ,eta=5,Rs=0,weights=None);G1_5_0
```

```python
fig3 = SimilarityMaps.GetSimilarityMapFromWeights(mH, G1_5_0, colorMap='jet',contourLines=10,scale=-1)
```

## Calculate the Symmetry Function 'G1' with the weights set to be charge

- G1 scaled by 1 $eta$=2, and $Rs$=0

```python
G1_q_2_0 = sum_G1(XYZ, eta=2, Rs=0, weights=charge);G1_q_2_0
```

```python
fig4 = SimilarityMaps.GetSimilarityMapFromWeights(mH, G1_q_2_0, colorMap='jet',contourLines=10,scale=-1)
```

```python
fig5 = SimilarityMaps.GetSimilarityMapFromWeights(mH, G1_q_5_0, colorMap='jet',contourLines=10,scale=-1)
```
