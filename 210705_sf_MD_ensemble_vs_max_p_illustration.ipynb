```python
fave_star = 'logp/DATASETS/sf_ave_star_MD_weights_charge.csv'
fmax_star = 'logp/DATASETS/sf_max_star_MD_weights_charge.csv'

results_ave_star = load_pd(fave_star)
results_max_star = load_pd(fmax_star)
```

```python
fave_Huuskonen = 'logp/DATASETS/sf_ave_Huuskonen_MD_weights_charge.csv'
fmax_Huuskonen = 'logp/DATASETS/sf_max_Huuskonen_MD_weights_charge.csv'

results_ave_Huuskonen = load_pd(fave_Huuskonen)
results_max_Huuskonen = load_pd(fmax_Huuskonen)
```

```python
f_name_star = 'logp/DATASETS/MD/star/md_star_CHON.csv'

d_star = pd.read_csv(f_name_star)
smiles_star = d_star['smile'].values
```

```python
# get the fps of the star datasets
fps_ave_star_0 = results_ave_star[4][2]
fps_max_star_0 = results_max_star[4][2]

smile_star_0 = smiles_star[4]
```

```python
mol = Chem.MolFromSmiles(smile_star_0)
mH = Chem.AddHs(mol)
```

```python
fig1 = SimilarityMaps.GetSimilarityMapFromWeights(mH, fps_ave_star_0[:,23], colorMap=None,contourLines=10,scale=-1)
```

```python
fig2 = SimilarityMaps.GetSimilarityMapFromWeights(mH, fps_max_star_0[:,23], colorMap=None,contourLines=10,scale=-1)
```

```python
fave_physprop = 'logp/DATASETS/sf_ave_PhysProp_MD_weights_charge.csv'
fmax_physprop = 'logp/DATASETS/sf_max_PhysProp_MD_weights_charge.csv'

d_ave_physprop = load_pd(fave_physprop)
d_max_physprop = load_pd(fmax_physprop)
```
